## Data Pre-processing

```python
array = data.to_numpy()
# print('array :=>\n ', array, '\n' + '-'*80, '\n')
X = array[:, 0:13]
# print('X :=>\n ', X, '\n' + '-'*80, '\n')
y = array[:, 13]
# print('y :=>\n ', y, '\n' + '-'*80, '\n')
```

## Data Classification and Model Selection

```python
kfold = KFold(n_splits=10, shuffle=True, random_state=7)
# print('kfold :=>\n ', kfold, '\n' + '-'*80, '\n')
cart_model = DecisionTreeRegressor(random_state=10, criterion='poisson')
# cart_model = DecisionTreeRegressor(random_state=10, criterion='friedman_mse')
# print('cart_model :=>\n ', cart_model, '\n' + '-'*80, '\n')
```

## Data Scrubbing and Error Estimation

```python
scoring = 'neg_mean_squared_error'
results = cross_val_score(cart_model, X, y, scoring=scoring, cv=kfold)
# print('results :=>\n ', results, '\n' + '-'*80, '\n')
print('results.mean() :=>\n ', results.mean(), '\n' + '-'*80, '\n')
```
