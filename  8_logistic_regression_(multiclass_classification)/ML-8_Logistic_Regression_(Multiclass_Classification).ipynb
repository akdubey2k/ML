# Logistic Regression: Multiclass Classification
In this tutorial we will see how to use logistic regression for multiclass classification.

```python
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split 
```

```python
digits = load_digits()
digits
```

```python
%matplotlib inline
plt.gray()
for i in range(3):
    plt.matshow(digits.images[i])
```

```python
dir(digits)
```

```python
digits.data[0:3]
```

```python
digits.target[0:3]
```

```python
digits.target_names[0:3]
```

#### Create and train the <font color='maroon'>dataset</font> using <font color='maroon'>Logistic Regression model</model>

```python
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
model.fit(X_train, y_train)
```

```python
len(X_train)
```

```python
len(X_test)
```

#### Measure accuracy of our model

```python
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[102])
```

```python
digits.target[102]
```

```python
# model.predict(digits.data[[102]]) # working fine
model.predict([digits.data[102]])
```

```python
model.predict(digits.data[0:3])
```

#### Confusion matrix

```python
y_predicted = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted) # confusion_matrix(y_true, y_pred, ...
cm
```

#### Re-present in a graphical way of format using <font color='maroon'>'seaborn'</font> 

```python
import seaborn as sb
plt.figure(figsize = (10, 8))
sb.heatmap(cm, annot=True)
plt.xlabel('Predicted', fontsize=16)
plt.ylabel('True', fontsize=16)
```
