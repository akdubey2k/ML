# 1. Definitions

## 1.1 Linear Regression
Linear regression is a statistical **supervised learning algorithm** used to model the linear relationship between a **dependent variable and one or more independent variables.**
<br>
The core idea is to establish a **linear equation** that can best calculate (predicts) the **dependent variable** based on the values of the **independent variables.**
<br>
<br>
### 1.1.1 Key Components
**Dependent Variable (Y):** Also called as **response or target** variable, that needs to be calculate (predict).<br>
**Independent Variables (X):** Also called as **predictors or features,** these are used to make calculation (predictions) about the dependent variable.
<br>
<br>
### 1.1.2 Linear Regression Equation:
#### ùëå = ùõΩ0 + ùõΩ1ùëã + ùúñ
* ùëå is the dependent variable.
* ùëã is the independent variable.
* ùõΩ0 is the intercept (the value of ùëå when ùëã is 0).
* ùõΩ1 is the slope (the change in ùëå for a one-unit change in ùëã).
* ùúñ is the residual part or error term (the difference between the **observed** and **predicted** values of **ùëå**).

## 1.2 Slope
The slope of a line is defined as the **ratio of the change in the ùë¶ coordinates to the change in the ùë• coordinates** between two points on the line.
<br>
If you have two points **(ùë•1, ùë¶1)** and **(ùë•2, ùë¶2)** on the line, the **slope ùëö** is calculated as:

##### ùëö = (ùë¶2 ‚àí ùë¶1) / (ùë•2 ‚àí ùë•1)

### 1.2.1 Slope-Intercept Form of a Line:
The slope is a key component in the slope-intercept form of a linear equation, which is written as:

**ùë¶ = ùëöùë• + ùëè**
<br>
Here:

* ùë¶ is the dependent variable.
* ùë• is the independent variable.
* ùëö is the slope of the line.
* ùëè is the y-intercept (the value of ùë¶ when ùë• = 0).

## 1.3 Ordinary Least Squares (OLS) / Mean Squared Error (MSE)
**Ordinary Least Squares (OLS)** is a method for estimating the **parameters (coefficients)** of a **linear regression model.** The goal of **OLS** is to find the **line (or hyperplane** in the case of multiple independent variables) that minimizes the sum of the squared differences between the **observed values** and the **values predicted** by the linear model. These differences are known as **residuals.**
<br><br>
### 1.3.1 Linear regression, model is based on **ordinary least square (OLS) / Mean square errror (MSE)**

### 1.3.4 Key Points:
* **OLS:** Used to **estimate the parameters** of the linear regression model.
* **MSE:** Used to **evaluate the performance** of the model by **averaging the squared residuals (errors).**

### 1.3.5 Conclusion
**OLS and MSE** are fundamental concepts in linear regression.
* **OLS** provides a method to find the **best-fitting line** by minimizing the **sum of squared residuals,** while
* **MSE** evaluates the **model's performance by averaging these squared residuals.**

*Both are crucial for understanding and applying linear regression effectively.*

### 1.3.6 Hypothetical representation
Consider;
* **Xi** to denote the independent variable
* **Yi** to denote dependent variable.
* A pair of **(Xi, Yi)** is called training example.
* The subscripe **`i`** in the notation is simply index into the training set.
* **`m`** number of training example then i = 1, 2, 3, . . . m
<br>

The goal of supervised learning is to learn a hypothesis function **`[ f ]`**, for a given training set, that can used to estimate **`Y based on X`**. So hypothesis fuction represented as

**`fŒ∏(Xi) = Œ∏0 + Œ∏1Xi`**

* **Œ∏0, Œ∏1** are parameter of hypothesis. This is an equation for Simple / Univariate Linear regression.

## 1.4 Steps to perform Linear Regression:
* **Collect Data:** Gather data for *the dependent variable and the independent variable(s).*
* **Fit the Model:** Use statistical techniques to estimate the coefficients **(ùõΩ)** that minimize the difference between the predicted and actual values of **ùëå.**
* **Evaluate the Model:** Assess the goodness of **fit** using metrics such as *R-squared, adjusted R-squared,* and the standard error of the estimate.
* **Make Predictions:** Use the fitted model to make predictions for new data.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

## 3. Dataset Loading

```python
# load the training dataset
url = "https://raw.githubusercontent.com/akdubey2k/ML/main/ML_1_Simple_Linear_Regression/ML_1_train.csv"
df = pd.read_csv(url)
df.head(5)
# df
```

## 5. Data Cleaning

```python
# clean the null data by dropping it from the dataset.
# It will remove a row from both 'dependent and independent' feature value.
df.dropna(inplace=True)
```

## 6. Dataset splitting, into training and test (validation) set

```python
# Splitting the dataset into training and test sets
#                                                  independent,dependent
X_train, X_test, y_train, y_test = train_test_split(df[['x']], df['y'], test_size=0.2, random_state=4)
```

```python
print("\nindependent training dataset : \n", X_train)
print("\ndependent training dataset : \n", y_train)
print("\nindependent test (validation) dataset : \n", X_test)
print("\ndependent test (validation) dataset : \n", y_test)
```

## 7. Model Creation

```python
# Creating a linear regression model
lr_model = LinearRegression()
```

## 8. Model Training

```python
# Fit the model to the training dataset
lr_model.fit(X_train, y_train)
```

```python
# ùëå = ùõΩ0 + ùõΩ1ùëã + œµ
# coefficient (ùõΩ1(beta) = slope) of linear regression model
coef = lr_model.coef_
coef
```

```python
# (ùõΩ0(beta) = y-intercept constant, when X is 0) of linear regression model
intercept = lr_model.intercept_
intercept
```

### ùëå = ùõΩ0 + ùõΩ1ùëã + œµ
### *y_predicted = X_test * coefficient + intercept*

```python
X_test.iloc[0] * coef + intercept
```

```python
X_test.iloc[1] * coef + intercept
```

```python
X_test.iloc[2] * coef + intercept
```

## 9. Model Prediction

```python
y_pred = lr_model.predict(X_test)
y_pred
```

##### *Comparison between 'y_test' (Actual/observed) 'y_pred' (Predicted)*

```python
if len(y_test) != len(y_pred):
  print("\nThe lists 'y_true' and 'y_pred' must have the same length.\n")
else:
  # threshold = 2  # Define a threshold for the difference
  red_color = "\033[91m"
  reset_color = "\033[0m"

print(f"{'y_true':>15} {'y_pred':>15}")
print("="*31)

for true, pred in zip(y_test, y_pred):
  diff = abs(true - pred)
  threshold = 0.1 * abs(true)  # Define a threshold 10% of the true value for the difference
  if diff > threshold:
    print(f"{red_color}{true:>15.5f} {pred:>15.5f}{reset_color}")
  else:
    print(f"{true:>15.5f} {pred:>15.5f}")
```

## 10. Data Visualization

```python
# Plot the data point and the regression line for training (X_train, y_train) dataset.
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, lr_model.predict(X_train), color='blue') # plot regression line
plt.xlabel('X - independent features')
plt.ylabel('y - dependent features')
plt.title("Linear Regression for Training Dataset")
plt.show()
```
