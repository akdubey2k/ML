# Dataset Description
## Context
This dataset is originally from **"the National Institute of Diabetes and Digestive and Kidney Diseases"**<br>
The objective of the dataset is to diagnostically predict <u>*whether or not a patient has diabetes*</u>, based on certain diagnostic measurements included in the dataset.<br>

Several constraints were placed on the selection of these instances from a larger database.<br>
In particular, all patients here are <font color='violet'>**females at least 21 years old of Pima Indian heritage.**</font>

## Content
The datasets consists of several medical predictor <font color='grey'><u>**independent variables and one target dependent variable, Outcome.**</u></font><br>
Predictor variables includes
- the number of pregnancies the patient has had,
- their BMI,
- insulin level,
- age, and so on.

### Description of the dataset
* **Pregnancies:** Number of times pregnant (continuous variable)
* **Glucose:** Plasma glucose concentration a 2 hours in an oral glucose tolerance test (continuous variable)
* **BloodPressure:** Diastolic blood pressure (mm Hg) (continuous variable)
* **SkinThickness:** Triceps skin fold thickness (mm) (continuous variable)
* **Insulin:** 2-Hour serum insulin (mu U/ml) (continuous variable)
* **BMI:** Body mass index (weight in kg/(height in m)^2) (continuous variable)
* **DiabetesPedigreeFunction:** Diabetes pedigree function (continuous variable)
* **Age:** age (years) (continuous variable)
* **Outcome:** Class variable (0 or 1) 268 of 768 are 1, the others are 0 (0: No diabete, 1: diabete) (binary variable)<br>

*These variables are used to analyze risk factors and symptoms associated with diabete diseases.*
