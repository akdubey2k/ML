### Predicting if a person would buy life insurnace based on his age using logistic regression

Above is a binary logistic regression problem as there are only two possible outcomes (i.e. if person buys insurance or he/she doesn't).

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv('insurance_data.csv')
df.head()
```

```python
%matplotlib inline
plt.scatter(df.age,df.buy_insurance,marker='+',color='red')
plt.xlabel('age',fontsize=16)
plt.ylabel('insurance cover (1/0)',fontsize=16)
```

```python
df.shape # return (rows, columns)
```

#### Train and test the datasets using <font color='brown'>sklearn.model_selection.train_test_split</font> method

```python
X_train,X_test,y_train,y_test = train_test_split(df[['age']],df['buy_insurance'],train_size=0.9)
model = LogisticRegression()
```

```python
X_test
```

#### Train the model with latest dataset prepared by sklearn module
Not used by previous one like this __model.fit(df[['age']],df['buy_insurance'])__

```python
model.fit(X_train,y_train)
```

```python
y_predicted = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.predict_proba(X_test) # down array explanation as it is [not buy probability, buy probability]
```

__model.coef___ indicates value of __<font color='brown'>m in y=m*x + b</font>__ equation.

```python
coefficient = model.coef_
coefficient
```

__model.intercept___ indicates value of __<font color='brown'>b in y=m*x + b</font>__ equation

```python
intercept = model.intercept_
intercept
```

## Lets defined sigmoid function now and do the math with hand
### z = y
### z = m*x + b

```python
import math
def sigmoid(x):
    return (1 / (1 + math.exp(-x)))
```

```python
def prediction_fun(age):
    z = coefficient * age + intercept
    y = sigmoid(z)
    return y
```

```python
age = 55
prediction_fun(age)
```

__0.9431 is more than 0.5 which means person with 55 age will buy insurance__

age = 55 ==> y = coeff * age + inter | (sigmoid/logit) z = 1 / (1 + math.exp(-y)) ==> __<font color='green'>0.94<font>__ 

```python
prediction_fun(25)
```

__0.129 is less than 0.5 which means person with 25 age will not buy insurance__

age = 25 ==> y = coeff * age + inter | (sigmoid/logit) z = 1 / (1 + math.exp(-y)) ==> __<font color='red'>0.12<font>__
