# 1. Definition
## 1.2 Description
A **decision tree** is a type of algorithm used in decision-making processes and machine learning. It is a *flowchart-like structure or graphical representation,* used to make decisions or predictions.

## 1.2 Key components and features of a decision tree:
**1.	Root Node:** This is the top node of the tree and represents the entire dataset. It is the starting point of the decision-making process.

**2.	Decision Nodes:** These nodes represent the points where the dataset is split based on certain conditions or attributes. Each decision node leads to further nodes.

**3.	Leaf Nodes (Terminal Nodes):** These nodes represent the final outcome or decision of the decision tree. They do not split further and provide the result based on the conditions tested in the path leading to them.

**4.	Branches:** These are the connections between nodes, representing the flow from one decision node to another or to a leaf node.

**5.	Splitting:** This is the process of dividing a node into two or more sub-nodes based on a certain condition or attribute.

**6.	Pruning:** This involves removing parts of the tree that do not provide additional power in classifying instances. Pruning helps to reduce the complexity of the model and prevent overfitting.


## 1.3 Working of Decision Tree Algorithm
1. Select the best attribute using **Attribute Selection Measures (ASM)** to split the records.
2. Make that attribute a decision node and breaks the dataset into smaller subsets.
3. Start tree building by repeating this process recursively for each child until one of the conditions will match:
  3.1 All the tuples belong to the same attribute value.<br>
  3.2 There are no more remaining attributes.<br>
  3.3 There are no more instances.<br>

## 1.4 Attribute Selection Measures
* **Attribute selection measure** is a heuristic (*commonsense rule intended to increase the probability of solving problems*) for selecting the **splitting criterion** that **partitions data** in the best possible manner.

* It is also known as **splitting rules** because it helps us to determine breakpoints for tuples on a given node.

* **ASM** provides **a rank to each feature (or attribute)** by explaining the given dataset.

* The **best score attribute** will be selected as a **splitting attribute (Source).**
* In the case of a **continuous-valued attribute, split points for branches also need to define.**

* The most popular selection measures are
1. Information Gain,
2. Gain Ratio, and
3. Gini Index.

## 2. Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

## 3. Load Data
#### Pima Indians Diabetes Database
Predict the onset of diabetes based on diagnostic measures

```python
url = "https://raw.githubusercontent.com/akdubey2k/ML/main/ML_13_Decision_Tree_Diabetes_Database/ML_13_Decision_Tree_Diabetes_Database.csv"
df = pd.read_csv(url)
df.head()
```

## 4. Feature Selection
Divide given columns into two types :
* variables dependent(or target variable)
* independent variable(or feature variables)

```python
# split dataset in features and target variable
X = df.drop('Outcome', axis=1)
y = df['Outcome']
print(f"{'Shape of independent variable(or feature variables)':60} {X.shape}")
print(f"{'Shape of dependent variable(or target variables)':60} {y.shape}")
```

## 5. Split Dataset
Dividing the dataset into a **training set and a test set** using the function **train_test_split().**

```python
# Split dataset into 70% training set and 30% test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

## 6. Build the Model
### Create a "decision tree model" using *Scikit-learn.*

```python
# Create Decision Tree classifer object
dt_clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
dt_clf = dt_clf.fit(X_train, y_train)

# Predict the response for test dataset
y_pred = dt_clf.predict(X_test)
# print(f"Shape of prediction {'.' * (80 - len('y_pred.shape'))} {y_pred.shape}")
print("Shape of prediction ".ljust(80, '.'), y_pred.shape)
# print(f"Value{'.' * (60 - len('Value'))}")
print(f"{'Prediction values '} \n {y_pred}")
```

## 7. Evaluate the Model
To estimate the accuracy of the decision tree classifier or model.<br>
Compute and compare the **actual test set values and predicted values.**

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy ".ljust(60, '.'), ":", accuracy_score(y_test, y_pred))
```
