# Import Libraries

```python
from pandas import read_csv
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.naive_bayes import ComplementNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
data = read_csv('diabetes.csv')
# data.head()
```

## check if feature having null/na entity (row)

```python
# print("data.info() => \n", data.info(), "\n" + "-"*80) # right to left print call
# print("data.isnull().sum() => \n",data.isnull().sum(), "\n" + "-"*80)
# print("data.shape => \n", data.shape, "\n" + "-"*80)
```

## Data extraction from the pandas dataframe

```python
array = data.values
# print("array => \n", array, "\n" + "-"*80)
X = array[:, 0:8]
# print("X => \n", X, "\n" + "-"*80)
y = array[:, 8]
# print("y => \n", y, "\n" + "-"*80)
```

## Spot Check Classification Algorithm

```python
gnb_model = GaussianNB()
bnb_model = BernoulliNB()
cat_nb_model = CategoricalNB()
comp_nb_model = ComplementNB()
mnb_model = MultinomialNB()

kfold = KFold(n_splits=10, shuffle=True, random_state=1)

results = cross_val_score(gnb_model, X, y, cv=kfold)
print("GaussianNB Score mean() => \n", results.mean(), "\n" + "-"*80)

results = cross_val_score(bnb_model, X, y, cv=kfold)
print("BernoulliNB Score mean() => \n", results.mean(), "\n" + "-"*80)

# results = cross_val_score(cat_nb_model, X, y, cv=kfold)
# print("CategoricalNB Score mean() => \n", results.mean(), "\n" + "-"*80)

results = cross_val_score(comp_nb_model, X, y, cv=kfold)
print("ComplementNB Score mean() => \n", results.mean(), "\n" + "-"*80)

results = cross_val_score(mnb_model, X, y, cv=kfold)
print("MultinomialNB Score mean() => \n", results.mean(), "\n" + "-"*80)
```
