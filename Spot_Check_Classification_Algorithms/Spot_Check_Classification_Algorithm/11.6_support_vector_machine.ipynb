# Import Libraries

```python
from pandas import read_csv
from sklearn.svm import SVC
from sklearn.svm import LinearSVC
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
data = read_csv('diabetes.csv')
# data.head()
```

## check if feature having null/na entity (row)

```python
# print("data.info() => \n", data.info(), "\n" + "-"*80) # right to left print call
# print("data.isnull().sum() => \n",data.isnull().sum(), "\n" + "-"*80)
# print("data.shape => \n", data.shape, "\n" + "-"*80)
```

## Data extraction from the pandas dataframe

```python
array = data.values
# print("array => \n", array, "\n" + "-"*80)
X = array[:, 0:8]
# print("X => \n", X, "\n" + "-"*80)
y = array[:, 8]
# print("y => \n", y, "\n" + "-"*80)
```

## Spot Check Classification Algorithm

```python
svc_model = SVC()
# linear_svc_model = LinearSVC(dual='auto', random_state=1, tol=1e-5)
kfold = KFold(n_splits=20, shuffle=True, random_state=7)

results = cross_val_score(svc_model, X, y, cv=kfold)
print("SVM Score mean() => \n", results.mean(), "\n" + "-"*80)

# results = cross_val_score(linear_svc_model, X, y, cv=kfold)
# print("LinearSVC Score mean() => \n", results.mean(), "\n" + "-"*80)
```
