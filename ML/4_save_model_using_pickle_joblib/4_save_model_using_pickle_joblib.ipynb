```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('homepreices.csv')
df
```

```python
model = LinearRegression()
model.fit(df[['area']], df.price)
model.score(df[['area']], df.price)
```

```python
model.predict([[5000]])
```

```python
import pickle
```

```python
with open('pickle_model', 'wb') as f:
    pickle.dump(model, f)
```

```python
with open('pickle_model', 'rb') as f:
    p_model = pickle.load(f)
```

```python
p_model.predict([[5000]])
```

```python
import joblib
```

```python
joblib.dump(model, 'joblib_model')
```

```python
j_model = joblib.load('joblib_model')
```

```python
j_model.predict([[5000]])
```

```python
j_model.score(df[['area']], df.price)
```

```python
p_model.score(df[['area']], df.price)
```
