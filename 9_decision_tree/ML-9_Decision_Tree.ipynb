```python
import pandas as pd
df = pd.read_csv("salaries.csv")
df.head()
```

#### Create input and target variables

```python
inputs = df.drop("salary_more_then_100k", axis='columns')
inputs.head()
```

```python
target = df['salary_more_then_100k']
target.head()
```

#### Create label encoder for the string variables, since ML do not recognise string. OneHotEncode is not suitable here for decision tree

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

#### Fit and transformer the features (independent variable)

```python
# this will automatically add the new column in the same matrix

inputs['n_company'] = le_company.fit_transform(inputs['company'])
inputs['n_job'] = le_job.fit_transform(inputs['job'])
inputs['n_degree'] = le_degree.fit_transform(inputs['degree'])
```

#### Get to know encoded values assigned by Label encoding 

```python
print(le_company.classes_) # it begins from zero
print(le_job.classes_)
print(le_degree.classes_)
```

#### Drop all the previous columns, which are remained 

```python
inputs.head()
```

```python
inputs_n = inputs.drop(['company', 'job', 'degree'], axis='columns')
inputs_n.head()
```

#### Use tree for decision 

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

#### Train the decision tree 

```python
model.fit(inputs_n, target)
```

#### Is salary of Google, Computer Engineer, Bachelors degree > 100 k ?

```python
model.predict([[2,1,0]])
```

#### Is salary of Google, Computer Engineer, Masters degree > 100 k ?

```python
model.predict([[2,1,1]])
```
